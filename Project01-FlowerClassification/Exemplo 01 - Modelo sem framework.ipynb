# Modelo de Aprendizado para Gatos e Cachorros

```python
# Cada animal tem duas características:
# x1 = comprimento (cm), x2 = altura (cm)

# Gatos (classe 0)
gatos = [
    (20, 25),  # comprimento, altura
    (22, 24),
    (19, 23),
]

# Cachorros (classe 1)
cachorros = [
    (40, 45),
    (42, 47),
    (39, 44),
]

# Organizar os dados
animais = gatos + cachorros
classificados = [0, 0, 0, 1, 1, 1]  # 0 para gato, 1 para cachorro
```

```python
# Pesos para comprimento e altura
w1 = 1.0  # peso para comprimento
w2 = 1.0  # peso para altura
b = -60   # bias
```

```python
# Função de soma (modelo linear)
def soma(x1, x2):
    return w1 * x1 + w2 * x2 + b

# Função de ativação (step function)
def ativacao(valor_soma):
    if valor_soma >= 0:
        return 1  # cachorro
    else:
        return 0  # gato
```

```python
for i in range(len(animais)):
    x1, x2 = animais[i]
    resultado_soma = soma(x1, x2)
    classe_predita = ativacao(resultado_soma)
    print(f"Comprimento: {x1} cm, Altura: {x2} cm -> Classe prevista: {classe_predita} (Classe real: {classificados[i]})")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Separar para plotar
x1_gatos = [g[0] for g in gatos]
x2_gatos = [g[1] for g in gatos]

x1_cachorros = [c[0] for c in cachorros]
x2_cachorros = [c[1] for c in cachorros]

plt.scatter(x1_gatos, x2_gatos, color='blue', label='Gato')
plt.scatter(x1_cachorros, x2_cachorros, color='red', label='Cachorro')

# Gerar linha de separação (reta de decisão)
x_vals = np.linspace(15, 50, 100)
y_vals = (-w1*x_vals - b)/w2  # reorganizando: y = -(w1/w2)*x - (b/w2)

plt.plot(x_vals, y_vals, '--k', label='Fronteira de decisão')

plt.xlabel('Comprimento (cm)')
plt.ylabel('Altura (cm)')
plt.legend()
plt.title('Classificação de Gatos e Cachorros - Modelo Linear')
plt.grid(True)
plt.show()
```

# Verificando se são Gatos e Cachorros

```python
# Novos animais (comprimento, altura)
novos_animais = [
    (21, 24),  # provavelmente um gato
    (41, 46),  # provavelmente um cachorro
    (23, 27),  # provavelmente um gato
    (38, 43),  # provavelmente um cachorro
]
```

```python
for animal in novos_animais:
    x1, x2 = animal
    resultado_soma = soma(x1, x2)
    classe_predita = ativacao(resultado_soma)
    
    # Mostrar o resultado
    classe_nome = 'Cachorro' if classe_predita == 1 else 'Gato'
    print(f"Novo Animal -> Comprimento: {x1} cm, Altura: {x2} cm -> Classe prevista: {classe_nome}")
```

```python
x1_novos = [a[0] for a in novos_animais]
x2_novos = [a[1] for a in novos_animais]

plt.scatter(x1_novos, x2_novos, color='green', marker='x', label='Novos animais')
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Dados dos animais antigos
x1_gatos = [g[0] for g in gatos]
x2_gatos = [g[1] for g in gatos]

x1_cachorros = [c[0] for c in cachorros]
x2_cachorros = [c[1] for c in cachorros]

# Dados dos novos animais
x1_novos = [a[0] for a in novos_animais]
x2_novos = [a[1] for a in novos_animais]

# Criar o gráfico
plt.scatter(x1_gatos, x2_gatos, color='blue', label='Gato (treino)')
plt.scatter(x1_cachorros, x2_cachorros, color='red', label='Cachorro (treino)')
plt.scatter(x1_novos, x2_novos, color='green', marker='x', label='Novos animais')

# Gerar valores de x para desenhar a reta
x_vals = np.linspace(15, 50, 100)  # Valores de x de 15 até 50

# Usar a fórmula da reta
y_vals = (-w1 * x_vals - b) / w2  # Vem da equação: w1*x1 + w2*x2 + b = 0

# Desenhar a reta
plt.plot(x_vals, y_vals, '--k', label='Fronteira de decisão')

# Finalizar o gráfico
plt.xlabel('Comprimento (cm)')
plt.ylabel('Altura (cm)')
plt.legend()
plt.title('Classificação de Gatos e Cachorros - Modelo Linear')
plt.grid(True)
plt.show()
```
